```python
def addChild(tree,child,parent,i):
    i = i + 1
    tree.create_node(child,i,parent,data=1)
    
    return i 

def getLabel(tree,succs):
    res = list()
    for s in succs:
        res.append(tree.get_node(s).tag)

    return res

def getNodeByLabel(tree,label,ids):
    for id in ids:
        n = tree.get_node(id)
        if(label == n.tag):
            return n


def addTransaction(tree,parent,trans,i):
    for t in trans:
        #print(t," node: ",parent)
    
        children = getChilds(tree,parent)
        lbs = getLabel(tree,children)

        #print("Child of ", t, lbs)

        if(t in lbs):
            #updateData(tree,node)
            parent = getNodeByLabel(tree,t,children)
            updateData(tree,parent)
            #print("update")

        else:
            id = addChild(tree,t,parent,i)
            node = tree.get_node(id)
            i = id
            parent = node
        
        

    return i





def updateData(tree,node):
    node.data = node.data + 1


def getChilds(tree,node):
    #print(node)
    return node.successors(tree.identifier)


def createTree(transactions):
    i=0
    tree  = Tree()

    tree.create_node("root",0)

    root = tree.get_node(0)

    for t in transactions:

        if(len(t)>0):
            #print("current trans: ",t)
            
            first = t[0]

            children = getChilds(tree,root)
            lbs = getLabel(tree,children)



            #print("child of root",lbs)
            if(first in lbs):
                #print("path available")
                parent = getNodeByLabel(tree,first,children)
                updateData(tree,parent)
                i = addTransaction(tree,parent,t[1:],i)
            else:
                i = addTransaction(tree,root,t,i)

            #tree.show()

    return tree
```
